```python
import csv
import networkx as nx
import numpy as np
from operator import itemgetter 
from scipy.stats import kendalltau
import pandas as pd
import math
from numpy import linalg as LA

G=nx.Graph()

f = open("ca-netscience.txt", "r")
edge1=[]
edge2=[]
nodelist=[]
for x in f:
  e1,e2=x.split()
  edge1.append(int(e1)-1)
  edge2.append(int(e2)-1)
  for i in (int(e1)-1,int(e2)-1):
      if i not in nodelist:
          nodelist.append(i)
nodelist.sort()
t=0
nnodelist=[]
dic={}
for i in nodelist:
    dic[i]=t
    nnodelist.append(dic[i])
    t=t+1

nodelist=nnodelist
for  i in range(0, len(edge1)):
    G.add_edge(dic[edge1[i]], dic[edge2[i]])

nedge1=[]
nedge2=[]
for  i in range(0, len(edge1)):
   nedge1.append(dic[edge1[i]])
   nedge2.append(dic[edge2[i]])
   
edge1=nedge1
edge2=nedge2
        
n=len(nodelist)
d=[0]*n
for i in  range(0,n):
    d[i]=edge1.count(i)+edge2.count(i)

B=nx.betweenness_centrality(G)
    

nodes = []
values = []

for item in B.items():
    nodes.append(item[0]), values.append(item[1])
 
# printing keys and values separately

#print("befored sorted")
#print("values : ", values)
#print("Nodes : ", nodes)

values,nodes=zip(*sorted(zip(values,nodes)))
values,nodes=list(values),list(nodes)
values.reverse()
nodes.reverse()
#print("after sorted")
print("Values: ", values)
print("Nodes: ", nodes)
#print("Top-10 Values: ",values[:10])
print("Top-10 Nodes: ",nodes[:10])
    
    
# n=10
# x=[]
# BC=[]
# #print(len(G.nodes))
# for i in range(0,round(len(G.nodes())*0.75),n):
#     new_G=nx.Graph()
#     new_G = G.copy()
#     new_G.remove_nodes_from(nodes[:i])
#     #print(G.nodes)
#     #print("Gnodes",len(G.nodes))
#     #print("newGnodes",len(new_G.nodes))
#     x.append(i)
#     BC.append(nx.number_connected_components(new_G))
#     #print(i)
# print(x)
# print(BC)
# #print(len(nodes[:i]))

# df=pd.DataFrame(list(zip(x,BC)),columns=['No_of_Nodes_Removed','No_of_Connected_Components'])
# print(df)
# df.to_excel('US_BC_No_Connected_Components.xlsx','Sheet1')


# df1=pd.DataFrame(list(zip(values,nodes)),columns=['Values','Nodes'])
# print(df1)
# df1.to_excel('US_BC_VN.xlsx','Sheet1')
```
